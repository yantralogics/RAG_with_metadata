```python
from docx import Document
import os

# Function to extract metadata from a Word document
def extract_metadata(docx_file):
    doc = Document(docx_file)

    # Extract file name
    file_name = os.path.basename(docx_file)

    # Extract paragraphs
    paragraphs = [p.text for p in doc.paragraphs]

    # Extract headings
    headings = []
    for paragraph in doc.paragraphs:
        if paragraph.style.name.startswith('Heading'):
            headings.append(paragraph.text)

    # Extract page numbers (not directly supported in python-docx)
    # You may need to use additional libraries or techniques to get page numbers.

    return {
        "file_name": file_name,
        "paragraphs": paragraphs,
        "headings": headings
        # Add more metadata as needed
    }

# Example usage
if __name__ == "__main__":
    # Replace 'your_word_document.docx' with the path to your Word document
    docx_file = 'temp_data/output_V2.docx'
    
    metadata = extract_metadata(docx_file)
    print("File Name:", metadata["file_name"])
    print("Paragraphs:", metadata["paragraphs"])
    print("Headings:", metadata["headings"])
```

```python
from docx import Document
import json

def read_docx(filename):
    doc = Document(filename)
    data = []

    for paragraph in doc.paragraphs:
        item = {'type': 'paragraph', 'content': paragraph.text}
        data.append(item)

    for section in doc.sections:
        item = {'type': 'section', 'page_number': section.start_page_number}
        data.append(item)

    for paragraph in doc.paragraphs:
        if paragraph.style.name.startswith('Heading'):
            item = {'type': 'heading', 'level': int(paragraph.style.name[-1]), 'content': paragraph.text}
            data.append(item)

    return data

def convert_to_json(filename, json_filename):
    data = read_docx(filename)
    with open(json_filename, 'w') as f:
        json.dump(data, f, indent=4)

# Usage example
convert_to_json('temp_data/output_V2.docx', 'temp_data/output.json')
```

```python
def parse_section(section):
    return {
        'title': section.find('span', class_='mw-headline').get_text(strip=True),
        'content': [p.get_text(strip=True) for p in section.find_all('p')],
        'subsections': [parse_section(subsection) for subsection in section.find_all('div', class_='mw-parser-output')]
    }

def fetch_wikipedia_article(title):
    url = f"https://en.wikipedia.org/wiki/{title.replace(' ', '_')}"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        content_div = soup.find('div', id='mw-content-text')
        return parse_section(content_div)
    else:
        return None

# Usage
title = 'Python (programming language)'
article = fetch_wikipedia_article(title)
```
